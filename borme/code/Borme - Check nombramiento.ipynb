```python
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
# set database
db = client.borme
#set collection
facts = db.facts
```

```python

c = set()
```

```python

for fact in facts.find({"Nombramientos" : {"$ne":None}}):
    #querySQL = 'insert into borme.registros values (' + fact["Fecha"] + ',' + fact['Registro'] +',' + fact['Provincia']+ ');'
    if (100*fact["blanks"]/fact["characters"] < 22):
        if ':' in fact["Nombramientos"]:
            for b in fact["Nombramientos"].split(':')[0:-1]:
                if len(b.split('. ')[-1]) < 50:
                    c.add(b.split('. ')[-1].upper().strip())
 
```
